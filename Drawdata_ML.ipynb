```python
# 1️⃣ Instal·lació de la llibreria
!pip install drawdata scikit-learn matplotlib ipywidgets

# 2️⃣ Imports
from drawdata import ScatterWidget
import ipywidgets
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, HTML
from sklearn.linear_model import LogisticRegression
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
# 3️⃣ Crear els widgets
widget = ScatterWidget()
output = ipywidgets.Output()
```

```python
@output.capture(clear_output=True)
def on_change(change):
    df = widget.data_as_pandas
    if len(df) and df['color'].nunique() > 1:
        X = df[['x', 'y']].values
        y = df['color']

        # Entrenem el model
        classifier = LogisticRegression().fit(X, y)

        # Mostrem la frontera de decisió
        display(HTML("<br><br><br>"))
        fig = plt.figure(figsize=(8, 6))
        disp = DecisionBoundaryDisplay.from_estimator(
            classifier, X,
            response_method="predict_proba" if len(np.unique(df['color'])) == 2 else "predict",
            xlabel="x", ylabel="y",
            alpha=0.5
        )
        disp.ax_.scatter(X[:, 0], X[:, 1], c=y, edgecolor="k")
        plt.title(f"{classifier.__class__.__name__}")
        plt.show()
```

```python
# 5️⃣ Vinculem el widget amb la funció
widget.observe(on_change, names="data")
```

```python
# 6️⃣ Execució inicial (per mostrar-ho)
on_change(None)
display(ipywidgets.HBox([widget, output]))
```
