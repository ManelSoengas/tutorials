# **<font color='brown'>Machine Learning</font>**

---
Machine Learning (ML) es, básicamente, un conjunto de algoritmos matemáticos cuyo objetivo es aprender y analizar datos de manera automática, sin intervención humana. Busca relaciones y patrones entre los datos.

ML forma parte de una de las áreas de desarrollo de Inteligencia Artificial.

IA_ambits (4).png

ML se divide en cuatro categorías:




1.   **Aprendizaje supervisado** : El modelo es entrenado con datos etiquetados. El error entre la salida y la salida deseada es minimizado por el algoritmo. Utilizado en tareas de clasificación y regresión.
2.   **Aprendizaje no supervisado **: Los datos utilizados durante el entrenamiento no estan etiquetados. El algoritmo analiza los datos y los divide en función de sus cracerísticas (Clustering).

3.   **Aprendizaje semi-supervisado** : Se utilizan datos etiquetados y no etiquetados durante el entrenamiento. El procedimiento consiste en agrupar los datos en clusters mediante un algoritmo no supervisado y, a continuación, etiquetarlos utilizando los datos etiquetados.

4.   **Aprendizaje por refuerzo** : El algoritmo aprende a identificar qué acciones, mediante ensayo y error, acumulan una recompensa. Se utiliza en robótica y videojuegos.







En el desarrollo de proyectos de ML, se aplican metodologías que facilitan su desarrollo y aplicación. Estas metodologías dividen el proyecto en diferentes procesos, lo que proporciona una visión clara de las partes que intervienen en el mismo. Aunque existen diferentes metodologías, una de las más conocidas es CRISP-DM. Dependiendo del tipo de proyecto, su dificultad, el equipo humano o su alcance, pueden omitirse algunas fases o optar por otras metodologías.

En ML nos podemos encontrar con **diferentes tipos de problemas** a resolver:

1.  **La clasificación**. La variable objetivo en un problema de clasificación es categórica. Esto significa que el objetivo es asignar una entrada a una de las categorías o clases predefinidas. Por ejemplo, clasificar correos electrónicos como "spam" o "no spam", o predecir si un tumor es "benigno" o "maligno".
2.  **La regresión**. La variable objetivo en un problema de regresión es continua. El objetivo es predecir un valor numérico basado en las entradas. Por ejemplo, predecir el precio de una casa, la temperatura de un día o el número de ventas de un producto.

También hay que tener en cuneta el tipo de la clase objetivo:

1. **Binaria**: Dos posibles valores/clases (ej. "sí" o "no").
2. **Multiclase**: Más de dos valores/clases, una única clase por instancia (ej. "A", "B" o "C").
3. **Multietiqueta**: Más de dos valores/clases, múltiples clases posibles por instancia (ej. "A", "B" y "C" al mismo tiempo).

A partir de las definiciones se puede afirmar que estamos ante un problema de:

1.  Clasificación
2.  La clase objetivo es Multiclase, ya que admite tres valores posibles.
3.  Se trata de un aprendizaje supervisado, los datos de entrenamiento y test estan etiquetados.

## **<font color='brown'>Fase 6: Entrenamiento del modelo</font>**




---
Para cada uno de los algoritmos seleccionados, se ejecuta el procedimiento de entrenamiento. Cabe destacar que no se han ajustado los [hiperparámetros]( https://es.wikipedia.org/wiki/Hiperpar%C3%A1metro_(aprendizaje_autom%C3%A1tico) de los algoritmos. Con conjuntos de datos reales, suele ser necesario ajustar los hiperparámetros para lograr un comportamiento más eficiente del modelo. No es una tarea fácil y requiere procedimientos específicos para encontrar la configuración ideal.

Además, se añade información de una métrica (score = precision ) para darnos una idea del modelo que mejor se ajusta al aprndizaje de la dtección de la clase objetivo y su clasificación.

```python

for name, clf in zip(names, classifiers):
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    print(name +": " + str(score))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

for name, clf in zip(names, classifiers):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    print(f"{name} - Accuracy: {accuracy:.2f}, Precision: {precision:.2f}, Recall: {recall:.2f}, F1 Score: {f1:.2f}")
```

## **<font color='brown'>Métricas de evaluación</font>**

---

Para evaluar la calida de un modelo e snecesario disponer de unas métricas. Estás métricas cambian en función del área de la IA a desarrollar y/o del problema a resolver. En el campo del ML las más habituales són:

1. **Accuracy**: La proporción de predicciones correctas sobre el total de predicciones realizadas. Mide qué tan bien el modelo clasifica en general.

2. **Precision**: La proporción de verdaderos positivos sobre el total de positivos predichos. Indica cuántas de las predicciones positivas son realmente correctas.

3. **Recall**: La proporción de verdaderos positivos sobre el total de positivos reales. Mide la capacidad del modelo para encontrar todos los casos positivos.

4. **F1 Score**: La media armónica de precision y recall. Es útil cuando se busca un equilibrio entre precision y recall.

Cabe decir que en función del problema a resolver, puede ser interesante fijarse en una métrica determinada e intentar mejorarla si llega el caso.
