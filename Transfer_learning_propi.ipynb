# **Transfer Learning**

Por lo tanto, si se desea clasificar nuevos tipos de imágenes, es posible utilizar el modelo de red neuronal de aprendizaje profundo preentrenado como punto de partida para un modelo en la segunda tarea de interés, por ejemplo, con tus propios datos de imagen. Esto se denomina aprendizaje por transferencia. El aprendizaje por transferencia es popular porque permite entrenar redes neuronales de aprendizaje profundo con datos relativamente escasos. El potencial del aprendizaje por transferencia es enorme y puede realmente dar vida a la clasificación de imágenes.

```python
import os, zipfile, tarfile, pathlib
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Descarga del Dataset

url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
tgz_path = keras.utils.get_file("flower_photos.tgz", origin=url, extract=False)

data_root = pathlib.Path(tgz_path).with_suffix("")  # ruta sense .tgz
extract_dir = data_root.parent / "flower_photos"

if not extract_dir.exists():
    with tarfile.open(tgz_path, "r:gz") as tar:
        tar.extractall(path=data_root.parent)

base_dir = extract_dir  # conté subcarpetes: daisy, dandelion, roses, sunflowers, tulips
print("Subcarpetes (classes):", sorted(os.listdir(base_dir)))
```

```python
IMAGE_SIZE = 224
BATCH_SIZE = 32
VAL_SPLIT = 0.2 #el 20% de les dades es reservaran per validació

# Triem backbone: MobileNetV2 (pots canviar a VGG16/ResNet50/DenseNet/EfficientNet/Inception)
from tensorflow.keras.applications import mobilenet_v2
preprocess_input = mobilenet_v2.preprocess_input

#ImageDataGenerator fa augmentació de dades (rotacions, desplaçaments, zoom, flips horitzontals).
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=15, width_shift_range=0.1, height_shift_range=0.1,
    zoom_range=0.1, horizontal_flip=True,
    validation_split=VAL_SPLIT
)

train_gen = train_datagen.flow_from_directory(
    base_dir,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",      # <-- multiclasse
    subset="training",
    shuffle=True,
    seed=42
)
val_gen = train_datagen.flow_from_directory(
    base_dir,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="validation",
    shuffle=False
)

num_classes = train_gen.num_classes
print("Mapping de classes:", train_gen.class_indices)
print("num_classes =", num_classes)
```

```python
base = mobilenet_v2.MobileNetV2(
    weights="imagenet",
    include_top=False, #traiem la classificació original d’ImageNet
    input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3)
)
base.trainable = False  # fase de transfer: congela el backbone,
                        # congeles les seves capes per no malmetre els pesos apreso

inputs = keras.Input(shape=(IMAGE_SIZE, IMAGE_SIZE, 3))
x = base(inputs, training=False)           # BN en inference mentre està congelat
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(num_classes, activation="softmax")(x)  # multiclasse
model = keras.Model(inputs, outputs)

model.compile(
    optimizer=keras.optimizers.Adam(1e-3),
    loss="categorical_crossentropy",        # <-- multiclasse
    metrics=["accuracy", keras.metrics.TopKCategoricalAccuracy(k=3, name="top3")]
)

model.summary()
```

```python
# serveixen per controlar, optimitzar i registrar l’entrenament del model automàticament.
# Et permeten parar a temps, ajustar el learning rate, desar el millor model, etc
callbacks = [
    keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=3, restore_best_weights=True),
    keras.callbacks.ModelCheckpoint("flowers_transfer.keras", save_best_only=True, monitor="val_accuracy")
]

history = model.fit(
    train_gen,
    epochs=8,
    validation_data=val_gen,
    callbacks=callbacks
)

val_loss, val_acc, val_top3 = model.evaluate(val_gen, verbose=0)
print(f"[VAL][Transfer] loss={val_loss:.4f} acc={val_acc:.4f} top3={val_top3:.4f}")
```
